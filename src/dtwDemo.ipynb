```python
time = np.array(np.linspace(0, 4 * np.pi, 100))
sequence_x = np.array(np.sin(time))
sequence_y = np.array(np.sin(time/0.8 + np.pi/3))
for i in range(0, 100):
    sequence_x[i] = sequence_x[i]*(100-i)/100 + np.random.uniform(-0.1, 0.1) 
    sequence_y[i] = sequence_y[i]*(100-i)/100 + np.random.uniform(-0.1, 0.1) 
# sequence_x = np.array(np.sin(time))

print(len(sequence_x))

plt.plot(time, sequence_x, label='Sequence 1')
plt.plot(time, sequence_y, label='Sequence 2', linestyle='--', color='red')
plt.xticks([])
plt.yticks([])
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.show()
```
